```python
# First let's begin by importing all the libraries I'll likely be using during the assignment

import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics as mt
from sklearn import linear_model
import numpy.polynomial.polynomial as poly
```

### Q1.1:

```python
#Then let's import the train, test, and validation data.

train_data1_in = pd.read_csv('Dataset_1_train.csv', header=None, usecols=[0,1])
test_data1_in = pd.read_csv('Dataset_1_test.csv', header=None, usecols=[0,1])
validation_data1_in = pd.read_csv('Dataset_1_valid.csv', header=None, usecols=[0,1])

#Fit the training data to a 20 degree polynomial
z20 = poly.polyfit(train_data1_in[0], train_data1_in[1], 20)
xx = np.linspace(-1, 1, 100)
train_p20 = poly.polyval(xx, z20)

#Then try the model on the validation set, and calculate the mse in comparision to the expected values in the test set
validation_p20 = poly.polyval(validation_data1_in[0],z20)
MSE = mt.mean_squared_error(test_data1_in[1], validation_p20)
print("Test MSE: " + str(MSE))

#Plot the results
xxx= np.linspace(-1, 1, 50)
plt.plot(test_data1_in[0], test_data1_in[1], 'o', xxx, validation_p20,'-g', xx, train_p20, '-r', validation_data1_in[0], validation_data1_in[1], 'o')
plt.legend(['Test points', '20th degree poly - Validation', '20th degree poly - train', 'Validation Points'])
plt.axis([-1,1,-15,30])
plt.figure(num=None, figsize=(30, 30), dpi=320, facecolor='w', edgecolor='k')
plt.show()
```
